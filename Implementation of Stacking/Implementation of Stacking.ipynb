```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn import tree
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('diabetes.csv')
```

```python
x = data.drop(columns=['Outcome'])
y = data['Outcome']
```

```python
x
```

```python
y
```

## We use KNN and SVM Algorithms as our base Model

```python
train,val_train,test,val_test = train_test_split(x,y,test_size = 0.5,random_state = 355)
```

```python
x_train,x_test,y_train,y_test = train_test_split(train,test,test_size = 0.2,random_state = 355)
```

## KNN

```python
knn = KNeighborsClassifier()
knn.fit(x_train,y_train)
knn.score(x_test,y_test)
```

## SVM

```python
svm = SVC()
svm.fit(x_train,y_train)
svm.score(x_test,y_test)
```

### Let's get the prediction of all the base model on the validation set val_train

```python
prediction_val1 = knn.predict(val_train)
prediction_val2 = svm.predict(val_train)
```

```python
prediction_val1
```

```python
prediction_val2
```

### Let's get prediction  value for validation set together as 'predict_val'

```python
predic_val = np.column_stack((prediction_val1,prediction_val2))
```

```python
predic_val
```

### Let's the prediction for all the models on the test 'x_test'

```python
prediction_test1 = knn.predict(x_test)
prediction_test2 = svm.predict(x_test)
```

```python
prediction_test1
```

```python
prediction_test2
```

```python
predict_test = np.column_stack((prediction_test1,prediction_test2))
predict_test
```

## Random Forest

```python
rand_clf = RandomForestClassifier()
rand_clf.fit(predic_val,val_test)
```

```python
rand_clf.score(predict_test,y_test)
```

### we are tuning three hyperparameters right now, we are passing the different values for both parameters

```python
from sklearn.model_selection import train_test_split,GridSearchCV
```

```python
grid_param = {
    "n_estimators" : [90,100,115],
    'criterion': ['gini', 'entropy'],
    'min_samples_leaf' : [1,2,3,4,5],
    'min_samples_split': [4,5,6,7,8],
    'max_features' : ['auto','log2']
}
```

```python
grid_search = GridSearchCV(estimator=rand_clf,param_grid=grid_param,cv=5,n_jobs =-1,verbose = 3)
```

```python
grid_search.fit(predic_val,val_test)
```

```python
grid_search.best_params_
```
